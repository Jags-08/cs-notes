```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(bi_df['DB'],bi_df['Age'],color='c')
plt.title('DB vs Age')
plt.xlabel('DB')
plt.ylabel('Age')
plt.show()
```

```python
bi_df.plot(x = 'fNAME', y = 'Age', linewidth = 3, color = 'r')
```

```python
plt.figure(figsize = (5,5))
plt.scatter(x = 'fNAME', y = 'Age', data = bi_df)
plt.title('fName vs Age')
plt.xlabel('DB')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(bi_df.corr(numeric_only=True),annot = True, cmap = 'Blues', linewidths = 1)
```
