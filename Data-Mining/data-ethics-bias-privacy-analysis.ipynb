# **Pract 4 : Data Integration and Transformation**

Integrate multiple datasets and apply data transformation techniques such as normalization, aggregation, and discretization.

**1. Import Libraries**

```python
import pandas as pd
```

**2. Load the CSV Datasets**

```python
students = pd.read_csv("C:/Users/jagru/Downloads/DYPIU/DMW/LAB/students.csv")
marks = pd.read_csv("C:/Users/jagru/Downloads/DYPIU/DMW/LAB/marks.csv")
attendance = pd.read_csv("C:/Users/jagru/Downloads/DYPIU/DMW/LAB/attendance.csv")
demographics = pd.read_csv("C:/Users/jagru/Downloads/DYPIU/DMW/LAB/demographics.csv")
```

**3. Data Integration (Merge Datasets)**
Merge using StudentID

```python
data = pd.merge(students, marks, on="StudentID")
data = pd.merge(data, attendance, on="StudentID")
data = pd.merge(data, demographics, on="StudentID")

print("Integrated Dataset:")
print(data)
```

**4. Aggregation**

Calculate Total Marks and Average Marks

```python
data["TotalMarks"] = data["Math"] + data["Programming"] + data["Statistics"]
data["AverageMarks"] = data["TotalMarks"] / 3

print("\nDataset with Aggregation:")
print(data)
```

**5. Normalization (Min-Max)**

Normalize AverageMarks

```python
#converting values into a common scale (usually 0 to 1)
data["NormalizedMarks"] = (data["AverageMarks"] - data["AverageMarks"].min()) / \
                          (data["AverageMarks"].max() - data["AverageMarks"].min())

print("\nAfter Normalization:")
print(data[["StudentID","AverageMarks","NormalizedMarks"]])
```

**6. Discretization**

Convert AverageMarks → Grade

```python
def assign_grade(marks):
    if marks <= 60:
        return "Low"
    elif marks <= 80:
        return "Medium"
    else:
        return "High"

data["Grade"] = data["AverageMarks"].apply(assign_grade)

print("\nAfter Discretization:")
print(data[["StudentID","AverageMarks","Grade"]])
```

**7. Final Dataset**

```python
print("\nFinal Transformed Dataset:")
print(data)
```
