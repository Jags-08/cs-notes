```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\dypiu23\Downloads\indian_student_dirty_dataset_realistic.csv")
```

```python
print("Initial Dataset Preview")
df.head()
```

```python
print("\n Dataset info")
df.info()
```

```python
print("\n Missing Values (Before Cleaning)")
df.isnull().sum()
```

```python
#Replace blanks with NaN
df.replace("", np.nan, inplace = True)
```

```python
'''
Handle missing values
numerical cols - mean & median
categorical cols - mode
'''
```

```python
#numerical columns
num_cols = ["Age", "CGPA", "Attendance_Percentage", "Annual_Fees", "Year_of_Study", "Admission_Year"]
```

```python
for col in num_cols:
    df[col].fillna(df[col].median(), inplace = True)
```

```python
#categorical columns - object means the column have string + integers
cat_cols = df.select_dtypes(include = "object").columns
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace = True)
```

```python
#Remove noisy data
df.loc[(df["Age"] < 17) | (df["Age"] > 25), "Age"] = df["Age"].median()
df.loc[(df["CGPA"] < 0) | (df["CGPA"] > 10), "CGPA"] = df["CGPA"].mean()
df.loc[(df["Admission_Year"] < 2015) | (df["Admission_Year"] > 2025), "Admission_Year"] = df["Admission_Year"].median()
df.loc[(df["Annual_Fees"] < 0), "Annual_Fees"] = df["Annual_Fees"].median()
```

```python
#Fix inconsistent data
df["Gender"] = df["Gender"].str.lower().replace({
    "m": "Male", "male": "Male",
    "f": "Female", "female": "Female"
})
```

```python
df["City"] = df["City"].str.title()
df["State"] = df["State"].str.title()
df["State"] = df["State"].replace({"Up": "Uttar Pradesh"})
df["Course"] = df["Course"].str.strip().replace({
    "Btech": "B.Tech", "Bsc": "B.Sc", "mba": "MBA"
})
df["Department"] = df["Department"].replace({"AIDS": "AI&DS"})
df["Caste_Category"] = df["Caste_Category"].str.upper()

binary_map = {"Y": "Yes", "y": "Yes", "yes": "Yes",
              "N": "No", "n": "No", "no": "No"}

df["Scholarship_Status"] = df["Scholarship_Status"].replace(binary_map)
df["Hostel_Resident"] = df["Hostel_Resident"].replace(binary_map)
```

```python
df.to_csv("indian_student_dirty_dataset_realistic.csv", index = False)
print("Data Cleaned")
```
