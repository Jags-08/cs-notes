```python
def classify_data_type(series):
    if series.dtype == 'object':
        return "Qualitative"
    elif np.issubdtype(series.dtype, np.integer):
        return "Quantitative - Discrete"
    elif np.issubdtype(series.dtype, np.floating):
        return "Quantitative - Continuous"
    else:
        return "Unknown"
```

```python
def identify_scale(column):
    col = column.lower()
    
    if col in ["student name", "grade"]:
        if col == "grade":
            return "Ordinal"
        return "Nominal"
    
    if col in ["maths", "phys", "chem", "bio", "eng"]:
        return "Interval"
    
    if col in ["study hours (per week)", "sgpa", "sr no."]:
        return "Ratio"
```

```python
results = []

for col in df.columns:
    data_type = classify_data_type(df[col])
    scale = identify_scale(col)
    
    results.append([col, data_type, scale])
    
classification_df = pd.DataFrame(
    results,
    columns = ["Attribute", "Data Type", "Scale of Measurement"]
)
classification_df
```
