```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms
```

```python
img = cv2.imread('/content/DIP Lab2.avif')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print("RGB image Shape :",img_rgb.shape)
plt.imshow(img_rgb)
plt.title("Orignal RGb Image")
plt.axis('off')
```

```python
#convert to grayscale for simplicity
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print("Grayscale Image Shape : ",img_gray.shape)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
pixel_block = img_gray[100:110,100:110]

print("10 x 10 pixel matrix \n")
print(pixel_block)
```

```python
plt.imshow(pixel_block, cmap='gray')
plt.title("pixel matrix grid(10 x 10)")
plt.colorbar()
plt.axis('off')
```

```python
equalized_img = cv2.equalizeHist(img_gray)
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Orignal Grayscale")
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(equalized_img, cmap='gray')
plt.title(" Histogram Equalized Grayscale")
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(img_gray.ravel(),bins=25)
plt.title("Grayscale before Equalization")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(equalized_img.ravel(),bins=25)
plt.title("Equalized img")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
r , g , b = cv2.split(img_rgb)

plt.figure(figsize=(15, 8))
plt.subplot(3,2,1)
plt.imshow(r, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(3,2,2)
plt.hist(r.ravel(), bins=256)
plt.title("Red Channel Histogram")

plt.subplot(3,2,3)
plt.imshow(g, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(3,2,4)
plt.hist(g.ravel(), bins=256)
plt.title("Green Channel Histogram")

plt.subplot(3,2,5)
plt.imshow(b, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(3,2,6)
plt.hist(b.ravel(), bins=256)
plt.title("Blue Channel Histogram")

plt.tight_layout()
plt.show()
```

```python
# histogram equalization on each channel
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)

plt.figure(figsize=(15,8))

# red channel
plt.subplot(3,2,1)
plt.imshow(r, cmap='gray')
plt.title("red channel image")
plt.axis('off')

plt.subplot(3,2,2)
plt.hist(r.ravel(), bins=256)
plt.title("red channel histogram")

# green channel
plt.subplot(3,2,3)
plt.imshow(g, cmap='gray')
plt.title("green channel image")

plt.subplot(3,2,4)
plt.hist(g.ravel(), bins=256)
plt.title("green channel histogram")

# blue channel
plt.subplot(3,2,5)
plt.imshow(b, cmap='gray')
plt.title("Blue channel image")
plt.axis('off')

plt.subplot(3,2,6)
plt.hist(b.ravel(), bins=256)
plt.title("Blue channel histogram")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,4))
# #histogram before equalization
plt.subplot(1,2,1)
plt.hist(img_gray.ravel(),bins=256)
plt.title("grayscale histogram (before equalization)")
plt.xlabel("pixel intensity")
plt.ylabel("frequency")
#histogram after equalization
plt.subplot(1,2,2)
plt.hist(equalized_img.ravel(),bins=256)
plt.title("grayscale histogram (after equalization)")
plt.xlabel("pixel intensity")
plt.ylabel("frequency")
plt.tight_layout()
plt.show()
```
