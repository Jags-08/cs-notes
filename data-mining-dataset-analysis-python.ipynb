```python
import pandas as pd
```

```python
df = pd.read_csv('/content/student_admission_record_dirty.csv')
```

```python
df['City'] = df['City'].str.strip()
```

```python
df['City'] = df['City'].str.upper()
```

```python
df['City'] = df['City'].str.lower()
```

```python
df['City'] = df['City'].str.title()
```

```python
df['City'].unique()
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
```

```python
df['Admission Test Score'] = df['Admission Test Score'].fillna(df['Admission Test Score'].median())
```

```python
from numpy._core.defchararray import lower
Q1 = df['Admission Test Score'].quantile(0.25)
Q3 = df['Admission Test Score'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Admission Test Score'] >= lower_bound) & (df['Admission Test Score'] <= upper_bound)]
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['smooth_column'] = df['Age'].rolling(window=3).mean()
```

```python
df.isnull().sum()
```

```python
df['Admission Status'] = df['Admission Status'].fillna(df['Admission Status'].mode()[0])
```

```python
df['High School Percentage'] = df['High School Percentage'].fillna(df['High School Percentage'].median())
```

```python
from numpy._core.defchararray import lower
Q2 = df['High School Percentage'].quantile(0.25)
Q4 = df['High School Percentage'].quantile(0.75)
IQR1 = Q4 - Q2

lower_bound1 = Q2 - 1.5 * IQR1
upper_bound1 = Q4 + 1.5 * IQR1

df = df[(df['High School Percentage'] >= lower_bound1) & (df['High School Percentage'] <= upper_bound1)]
```

```python
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])
```
